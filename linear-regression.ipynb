```python
%matplotlib widget

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

# Generate Data

```python
jumlah_data = 100

y = np.array([i*0.1+np.random.randn() for i in range(jumlah_data)])
x = np.array([i*0.1 for i in range(jumlah_data)])
```

### Fungsi Linear 

```python
def fungsi_linier(x,gradien):
    y = gradien*x
    return y
```

1. menghitung persamaan linear awal

```python
x_prediksi = np.array([0, 10])
m_awal = 1
y_prediksi = fungsi_linier(x_prediksi, m_awal)
```

2. menghitung prediksi

```python
m_prediksi = m_awal
m_list_prediksi = []
y_list_prediksi = []
x_list_prediksi = []
learning_rate = 0.1
for i in range(1,jumlah_data):
    y_prediksi = fungsi_linier(x[i],m_prediksi)
    y_aktual = y[i]

    error = y_aktual-y_prediksi
    delta_m = learning_rate*error/x[i]
    m_prediksi = m_prediksi+delta_m

    m_list_prediksi.append(m_prediksi)
    y_list_prediksi.append(fungsi_linier(np.array([0,10]),m_prediksi))
    x_list_prediksi.append(np.array([0,10]))
```

3. visualisasi prediksi

```python
fig = plt.figure(figsize=(4,4))
line, = plt.plot([],[],"Red")

def animate(frame_num):
    x = x_list_prediksi[frame_num]
    y = y_list_prediksi[frame_num]
    line.set_data((x,y))
    return line

plt.scatter(x,y)
plt.axis([0,10,0,10])
anime = FuncAnimation(fig,animate,frames=100,interval=100,repeat=False)
plt.show()
```
